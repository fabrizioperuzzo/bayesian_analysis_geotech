```python
# PyMC3 for Bayesian Inference
import pymc3 as pm
from scipy import stats
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mu = 4
sigma = 2
model0 = pm.Model()
with model0:
    pm.Normal('X', mu, sd=sigma)

with model0:
    trace = pm.sample(10000, start=dict(X=2), step=pm.Metropolis())

X = trace.get_values('X')

# nota in automatico usa 4 chain forse perché l'ultima volta che l'ho usato ho messo 4 chain
```

```python
x = np.linspace(-4,12,1000)
y = stats.norm(mu,sigma).pdf(x)

fig,ax = plt.subplots(figsize=(8,3))

ax.plot(x,y,'r',lw=2)
sns.distplot(X, ax=ax)
ax.set_xlim(-4,12)
```

```python
fig, axes = plt.subplots(1,2,figsize=(8,2.5), squeeze=False)

pm.traceplot(trace,ax=axes)
axes[0,0].plot(x,y,'r',lw=0.5)
```

```python
dataset = sm.datasets.get_rdataset("Davis", "carData")
```

```python
data = dataset.data[dataset.data.sex == 'M']
```

```python
data = data[data.weight < 110]
```

```python
data.head(3)
```

```python
# import statsmodels.formula.api as smf
model = smf.ols("height ~ weight", data=data)
```

```python
result = model.fit()
```

```python
x = np.linspace(50,110,25)
y = result.predict({"weight":x})
fig, ax = plt.subplots(1,1, figsize=(8,3))
ax.plot(data.weight, data.height, 'or')
ax.plot(x,y,color="blue")
ax.set_xlabel("weight")
ax.set_ylabel("height")
```

```python
with pm.Model() as model:
    sigma = pm.Uniform('sigma', 0, 10)
    intercept =pm.Normal('intercept', 125, sd=30)
    beta = pm.Normal('beta', 0, sd=5)
    height_mu = intercept + beta * data.weight
    pm.Normal('height',mu=height_mu, sd=sigma, observed=data.height)
    pred_height_mu = intercept + beta * x
    predict_height = pm.Normal('predict_height', mu=pred_height_mu, sd=sigma, shape=len(x))
```

```python
model.vars
```

```python
with model:
    trace = pm.sample(draw=10000, step=pm.NUTS(), start=pm.find_MAP())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 4), squeeze=False)
pm.traceplot(trace, varnames=['intercept', 'beta'], ax=axes)
```

```python
intercept = trace.get_values('intercept').mean()
intercept
```

```python
beta = trace.get_values("beta").mean()
beta
```

```python
fig, ax = plt.subplots(figsize=(14,5))
# plotto 1500 linee variabili
for n in range (500,1700,1):
    intercept = trace.get_values('intercept')[n]
    beta = trace.get_values('beta')[n]
    ax.plot(x, intercept+beta*x, color='red', lw=0.25, alpha=0.05)
# plotto la linea media
intercept = trace.get_values("intercept").mean()
beta = trace.get_values("beta").mean()
ax.plot(x, intercept + beta * x, color='k', label="Mean Bayesian prediction")
ax.plot(data.weight, data.height, 'o', label='data')
ax.plot(x, y, '--', color="blue", label="OLS prediction")
ax.set_xlabel("weight")
ax.set_ylabel("height")
ax.legend(loc=0)
```

#### We add Female Gender to analysis

```python
data1 = dataset.data.copy()
data1 = data1[data1.weight < 110]
data1["sex"] = data1["sex"].apply(lambda x: 1 if x == "F" else 0)
data1.head(3)
```

```python
with pm.Model() as model1:
    intercept_mu,  intercept_sigma = 125, 30
    beta_mu, beta_sigma = 0, 5
    
    # define model variables
    intercept = pm.Normal('intercept', intercept_mu, sd=intercept_sigma, shape=2)
    beta = pm.Normal('beta', beta_mu, sd=beta_sigma, shape=2)
    error = pm.Uniform('error',0,10)
    
    #define 
    sex_idx = data1.sex.values
    height_mu = intercept[sex_idx] + beta[sex_idx] * data1.weight
    
    #define hight
    pm.Normal('height', mu=height_mu, sd=error, observed=data1.height)
```

```python
with model1:
    trace1 = pm.sample(5000, chains=1,step=pm.NUTS(), start=pm.find_MAP())
```

```python
pm.traceplot(trace1, figsize=(8, 6))
```

```python
intercept_m, intercept_f = trace1.get_values('intercept').mean(axis=0)
intercept_m, intercept_f
```

```python
beta_m, beta_f = trace1.get_values('beta').mean(axis=0)
beta_m, beta_f
```

```python
intercept_mf = trace1.get_values('intercept').mean()
beta_mf = trace1.get_values('beta').mean()

intercept_mf, beta_mf
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,5))
mask_m = data.sex == 0  # return true false
mask_f = data.sex == 1  # return true false
ax.plot(data.weight[mask_m], data.height[mask_m], 'o', color='steelblue', label='male', alpha=0.5)
ax.plot(data.weight[mask_f], data.height[mask_f], 'o', color="green", label="female", alpha=0.5)
x = np.linspace(35, 110, 50)
ax.plot(x, intercept_m +x *beta_m, color="steelblue", label="model male group")
ax.plot(x, intercept_f + x * beta_f, color="green", label="model female group")
ax.plot(x, intercept_mf + x * beta_mf, color="black", label="model both groups")
ax.set_xlabel("weight")
ax.set_ylabel("height")
ax.legend(loc=0)
```
